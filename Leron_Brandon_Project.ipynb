```
def plot_pc(col_names, vt, k):
    """ 
    Plot how much each column of our data contributes 
    to each principal component and labels the rows of V^T.
    """
    df = pd.DataFrame({'v':vt[k, :], 'Column names':col_names})
    chart = alt.Chart(df).mark_bar().encode(
        x='Column names',
        y='v',
        opacity=alt.value(0.7)
    ).configure_axis(
        labelFontSize=12,
        titleFontSize=14
    ).configure_axisX(
        labelAngle = 0
    ).properties(width = 2000)
    return chart

pca.explained_variance_

u,s,vt = np.linalg.svd(df_aligned, full_matrices=False)

first_pc = (df_aligned @ vt.T)[[0]].rename({0: 'pc1', 1: 'pc2'}, axis = 1)


pca_reg_model2 = LogisticRegression()
pca_reg_model2.fit(first_pc,school)
y_hat_pca = pca_reg_model2.predict(first_pc)

print("Multilinear Regression predicting school fraction incorrect:", 
      zero_one_loss(school, pca_reg_model2.predict(first_pc) > 0.5))

plot_pc(list(df_aligned.columns), vt, 0)
```
